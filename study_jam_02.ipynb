```python
import numpy as np
x=np.array([1,2,3])
y=np.array([4,5,6])
print(x)
print(y)
```

```python
#data types
z=np.int_([4,5,6])
a=np.float32([2,5,7])
print(z)
print(a)
```

```python
#arange
np.arange(10)
```

```python
#values
x=np.zeros((2,3))#assign array with 0
print(x)

x=np.ones((3,3))#assign array with 1
print(x)

x=np.random.random((2,3))#assign random values
print(x)
```

```python
z=np.arange(1000)
print(z)
print(z[2])
print(z[-2])
```

```python
z.shape=(10,10,10)#changes the shape of array
print(z)
```

```python
z.shape=(5,2,5,2,2,5)
print(z)
```

```python
z=np.arange(30)
print(z)
```

```python
z.shape=(-1,5)#chooses the dimension 
print(z)
```

```python
#operations
x=np.arange(10)
y=np.arange(10)
print(x+y)
print(x-y)
print(x*y)
```

```python
print(np.sin(x))
print(np.cos(y))
```

```python
#matrix multiplication
np.dot(x,y)
```

```python
x.shape=(5,2)
print(x)
print(x.sum())
print(x.min())
print(x.max())
print(x.max(axis=0))
print(x.sum(axis=1))
```
